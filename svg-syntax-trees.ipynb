```python
from pathlib import Path
import tabulate
from IPython.display import HTML, SVG, display

def show_svg(raw_svg):
    display(SVG(data=raw_svg))
```

```python
in_dir = Path('input')
out_dir = Path('output')

# ensure the in and out folders exist
if not in_dir.is_dir():
    in_dir.mkdir(exist_ok=True)
if not out_dir.is_dir():
    out_dir.mkdir(exist_ok=True)

in_file = in_dir / 'test_processed.tsv'
content = in_file.read_text(encoding='utf-8-sig')
```

```python
table = [line.split('\t') for line in content.split('\n')]
display(HTML(tabulate.tabulate(table, tablefmt='html')))
```

```python
from syntactic_analysis.analysis import generate_trees

tree, version_trees = generate_trees(content)
```

```python
show_svg(tree.print_svg())
```

```python
for t in version_trees:
    show_svg(t.print_svg())
```
